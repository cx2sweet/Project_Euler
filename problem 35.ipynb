The number, 197, is called a circular prime because all rotations of the digits: 197, 971, and 719, are themselves prime.

There are thirteen such primes below 100: 2, 3, 5, 7, 11, 13, 17, 31, 37, 71, 73, 79, and 97.

How many circular primes are there below one million?

```
import numpy as np
```

```
#finds all primes
pn = np.arange(2,1000000) #possible nums
primes=[]
FLAG = True
while FLAG:
    i = pn[0]
    primes.append(i)
    pn = pn[pn%i!=0]
    if len(pn)==0: FLAG = False
primes = np.array(primes)
p3 = primes


cprimes = [2]
#p3 = np.array([110,101,197,971,719])
FLAG = True
while FLAG:

    b = str(p3[0])
    combos = []
    for i in range(len(b)):
        combos.append(int(b[i:]+b[:i]))
    combos = np.array(combos)

    #if all numbers are odd, this is to save on time
    if np.all(combos%2!=0):
        #if a1, a2, and a3 are in p3, then add nums to list then remove from p3
        if np.all(np.isin(combos,p3)):
            for i in combos:
                cprimes.append(i)
    
    #removes combos from p3
    for i in combos:
        p3 = p3[p3!=i]

    if len(p3)==0: FLAG = False

print(len(set(cprimes)))
```
