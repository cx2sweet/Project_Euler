The series, 11 + 22 + 33 + ... + 1010 = 10405071317.

Find the last ten digits of the series, 11 + 22 + 33 + ... + 10001000.

```
a=21
print(a**a)
print(str(a**a)[-10:])
print(int(str(a**a)[-10:]))
```

```
max = 1001
sum = 0
for i in range(1,max):
    s = str(i**i)[-12:]
    sum += int(s)

print(str(sum)[-10:])


```
