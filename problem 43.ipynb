The number, 1406357289, is a 0 to 9 pandigital number because it is made up of each of the digits 0 to 9 in some order, but it also has a rather interesting sub-string divisibility property.

Let d1 be the 1st digit, d2 be the 2nd digit, and so on. In this way, we note the following:

d2d3d4=406 is divisible by 2  
d3d4d5=063 is divisible by 3  
d4d5d6=635 is divisible by 5  
d5d6d7=357 is divisible by 7  
d6d7d8=572 is divisible by 11  
d7d8d9=728 is divisible by 13  
d8d9d10=289 is divisible by 17  
Find the sum of all 0 to 9 pandigital numbers with this property.

```
class blank_num:
    d1='n'
    d2='n'
    d3='n'
    d4='n'
    d5='n'
    d6='n'
    d7='n'
    d8='n'
    d9='n'
    d10='n'

    def __init__(self,d810='nnn',test=False):
        self.d8 = d810[0]
        self.d9 = d810[1]
        self.d10 = d810[2]
        if test==True: self.set_test()
        

    def set_test(self):
        self.d1='1'
        self.d2='4'
        self.d3='0'
        self.d4='6'
        self.d5='3'
        self.d6='5'
        self.d7='7'
        self.d8='2'
        self.d9='8'
        self.d10='9'

    def get(self,a,b):
        all = self.d1+self.d2+self.d3+self.d4+self.d5+self.d6+self.d7+\
        self.d8+self.d9+self.d10
        return all[a-1:b]

    def check05(self):
        all = self.d1+self.d2+self.d3+self.d4+self.d5+self.d6+self.d7+\
        self.d8+self.d9+self.d10
        all=all.lstrip('n')
        sl = set(list(all))
        if ('0' in sl) and ('5' in sl): return False
        else: return True

    def check(self):
        all = self.d1+self.d2+self.d3+self.d4+self.d5+self.d6+self.d7+\
        self.d8+self.d9+self.d10
        all=all.lstrip('n')
        if len(all) > len(set(all)): return False
        else: return True        
        

test_num = blank_num(test=True)
#test_num.set_test()
test_num.get(7,10)
#test_num.check()
```

```
m17=[]
for i in range(1,59):
    x=17*i
    if len(str(x))<3:
        m17.append('0'+str(x))
    if len(set(str(x)))==3: m17.append(str(x))

for pos,i in enumerate(m17):
    sli = set(list(i)) #set(list(i))
    #removes any numbers with repetitions
    if len(sli)<3: m17.pop(pos)
    #remove number if it has both 0 and 5
    elif ('0' in sli) and ('5' in sli): m17.pop(pos)

step1=[]

for d8_10 in m17:
    num = blank_num(d8_10)
    step1.append(num)
    print(num.get(1,10))
```

```
m13=[]
for i in range(1,77):
    x=13*i
    if len(str(x))<3:
        m13.append('0'+str(x))
    if len(set(str(x)))==3: m13.append(str(x))

m13b=[]

for i in m13:
    sli = set(list(i)) #set(list(i))
    #removes any numbers with repetitions
    if len(sli)==3:  #if all digits are unique
        if (('0' in sli) and ('5' in sli))==False:  #if not using both 0 and 5
            m13b.append(i)

m13=m13b
step2=[]
for num in step1:
    d8=num.get(8,8)
    d9=num.get(9,9)
    for a in m13:
        if a[1]==d8 and a[2]==d9:
            num.d7=a[0]
            step2.append(num)

step2b=[]
for i in step2:
    if i.check() and i.check05(): step2b.append(i)

step2=step2b

for i in step2b:
    print(i.get(1,10))
```
