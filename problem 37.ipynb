The number 3797 has an interesting property. Being prime itself, it is possible to continuously remove digits from left to right, and remain prime at each stage: 3797, 797, 97, and 7. Similarly we can work from right to left: 3797, 379, 37, and 3.

Find the sum of the only eleven primes that are both truncatable from left to right and right to left.

NOTE: 2, 3, 5, and 7 are not considered to be truncatable primes.

```
import numpy as np
```

```
#finds all primes
pn = np.arange(2,1000000) #possible nums
primes=[]
FLAG = True
while FLAG:
    i = pn[0]
    primes.append(i)
    pn = pn[pn%i!=0]
    if len(pn)==0: FLAG = False
primes = np.array(primes)
#print('found primes')
#3797
possible_TPs = primes[4:]
TPs=[]

for a in possible_TPs:
    #print(a)
    is_TP = True
    b=a
    lis = list(str(b))
    if ('0' in lis)  or ('4' in lis) or ('6' in lis) or ('8' in lis):
        continue

    while b>=10:
        b=int(str(b)[1:]) #removes first digit
        if ~np.isin(b,primes): 
            is_TP=False
            break

    b=a
    while b>=10:
        b = int(str(b)[:-1])
        if ~np.isin(b,primes): 
            is_TP = False
            break
    print(a)
    if is_TP: TPs.append(a)        

print(TPs)
print(len(TPs))
print(sum(TPs))
```
