Triangle, pentagonal, and hexagonal numbers are generated by the following formulae:

Triangle	 	Tn=n(n+1)/2	 	1, 3, 6, 10, 15, ...
Pentagonal	 	Pn=n(3n−1)/2	 	1, 5, 12, 22, 35, ...
Hexagonal	 	Hn=n(2n−1)	 	1, 6, 15, 28, 45, ...
It can be verified that T285 = P165 = H143 = 40755.

Find the next triangle number that is also pentagonal and hexagonal.

```
import math as math

def make_Tn(n):
    return int(n*(n+1)/2)

def is_Pn(P):

    U = math.ceil(1/6+math.sqrt(2/3*(P+1/24)))
    L = math.floor(1/6+math.sqrt(2/3*(P+1/24)))

    if P==L*(3*L-1)/2 or P==U*(3*U-1)/2:
        return True
        #print(P,'is a Pent number')
    else: return False

def is_Hn(h):
    a=2
    b=-1
    c=-h
    U = math.ceil((1+math.sqrt(1 - 8*c))/(4))
    L = math.floor((1+math.sqrt(1 - 8*c))/(4))
    if h == U*(2*U-1) or h == L*(2*L-1): return True
    else: return False
```
